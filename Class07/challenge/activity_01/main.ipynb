# ATIVIDADE 1

Você recebeu um arquivo vendas_loja.csv com informações de vendas de uma loja.
O arquivo contém valores nulos e inconsistências em algumas colunas.

## 1. LENDO ARQUIVO

```python
import matplotlib.pyplot as plt
import pandas as pd

FILEPATH = 'https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/vendas_loja.csv'
ENCODING = 'utf-8'
SEPARATOR = ','

try:
    df = pd.read_csv(filepath_or_buffer=FILEPATH, sep=SEPARATOR, encoding=ENCODING)
    display(df)
except FileNotFoundError:
    display('File not found')
    exit()
```

## 2. TRATANDO OS DADOS

### 2.1. VERIFICANDO NULOS

```python
df.isnull().sum()
```

### 2.2. SUBSTITUINDO VENDAS COM VALORES NULO (VENDA INVÁLIDA) POR 0

```python
SALES_WITH_NULL_VALUES = 'Vendas'

clean_df = df.copy()
clean_df[SALES_WITH_NULL_VALUES] = clean_df[SALES_WITH_NULL_VALUES].fillna(0)
# clean_df = df.dropna(subset=[SALES_WITH_NULL_VALUES]).reset_index(drop=True)

display(clean_df)
```

### 2.3. SUBSTITUINDO REGIÃO NULA POR 'INDEFINIDA'

```python
REGION_WITH_NULL_VALUES = 'Regiao'

clean_df[REGION_WITH_NULL_VALUES] = clean_df[REGION_WITH_NULL_VALUES].fillna('Indefinida')

display(clean_df)
```

## 3. GRÁFICOS

### 3.1. TOTAL DE VENDAS POR CATEGORIA (BAR CHART)

```python
df_sales_by_category = clean_df.groupby('Categoria')['Vendas'].sum()
display(df_sales_by_category)
```

```python
KIND = 'bar'
COLOR = 'orange'
FIGSIZE = (8,5)
TITLE = 'Total de Vendas por Categoria'
XLABEL = 'Categoria'
YLABEL = 'Vendas (R$)'
ROTATION = 0

ENABLE_GRID = True
GRID_LINESTYLE = '--'
GRID_ALPHA = 0.6
```

```python
df_sales_by_category.plot(
    kind=KIND,
    color=COLOR,
    figsize=FIGSIZE,
    title=TITLE,
    xlabel=XLABEL,
    ylabel=YLABEL,
    rot=ROTATION
)
plt.tight_layout()
plt.grid(axis='y', linestyle=GRID_LINESTYLE, alpha=GRID_ALPHA)
plt.show()
```

### 3.2. PROPORÇÃO DE VENDAS POR REGIÃO (PIE CHART)

```python
df_sales_by_region = clean_df.groupby('Regiao')['Vendas'].sum()
display(df_sales_by_region)
```

```python
KIND = 'pie'
FIGSIZE = (8,5)
TITLE = 'Proporção de Vendas por Região'
YLABEL = ''
AUTOPCT = '%1.2f%%'
STARTANGLE = 90
```

```python
df_sales_by_region.plot(
    kind=KIND,
    figsize=FIGSIZE,
    title=TITLE,
    ylabel=YLABEL,
    autopct=AUTOPCT,
    startangle=STARTANGLE,
)
plt.tight_layout()
plt.show()
```

### 3.3. EVOLUÇÃO DAS VENDAS AO LONGO DOS MESES

```python
df_sales_by_month = clean_df.groupby('Mes')['Vendas'].sum()
display(df_sales_by_month)
```

```python
KIND = 'line'
FIGSIZE = (8,5)
TITLE = 'Evolução das Vendas ao Longo dos Meses'
LEGEND = True
XLABEL = 'Mês'
YLABEL = 'Total de Vendas (R$)'
ROTATION = 0
MARKER = 'o'

ENABLE_GRID = True
GRID_LINESTYLE = '--'
GRID_ALPHA = 0.6
```

```python
df_sales_by_month.plot(
    kind=KIND,
    figsize=FIGSIZE,
    title=TITLE,
    xlabel=XLABEL,
    ylabel=YLABEL,
    rot=ROTATION,
    marker=MARKER
)
plt.tight_layout()
plt.grid(ENABLE_GRID, linestyle=GRID_LINESTYLE, alpha=GRID_ALPHA)
plt.show()
```
