## 2. IMPORT LIBRARIES

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## 3. LOAD DATA FROM FILE

```python
FILEPATH = 'https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/alunos_cursos.csv'
ENCODING = 'utf-8'
SEPARATOR = ';'

# pd.set_option('display.max_rows', 5)

try:
    df = pd.read_csv(filepath_or_buffer=FILEPATH, sep=SEPARATOR, encoding=ENCODING)
except FileNotFoundError:
    display('File not found')
    exit()

df.info()
```

## 4. CHARTS

### 4.1. NUMBER OF STUDENTS BY COURSE (COUNTPLOT CHART)

```python
X_LABEL = 'Curso'
COLOR_PALETTE = 'pastel'
HUE = 'Curso'
LEGEND = False
TITLE = 'Students per Course'

sns.countplot(data=df, x=X_LABEL, palette=COLOR_PALETTE, hue=HUE, legend=LEGEND)
plt.title(label=TITLE)
plt.grid(True)
plt.show()
```

### 4.2. STUDENT'S GRADE AVERAGE BY COURSE

```python
X_LABEL = 'Nota'
Y_LABEL = 'Aluno'
COLOR_PALETTE = 'pastel'
HUE = 'Curso'
MARKER = 'o'
LEGEND = True
TITLE = 'Average grade of students per course'

sns.lineplot(data=df, x=X_LABEL, y=Y_LABEL, marker=MARKER, palette=COLOR_PALETTE, hue=HUE, legend=LEGEND)
plt.title(label=TITLE)
plt.grid(True)
plt.show()
```

### 4.3. STUDENT'S GRADE AVERAGE BY A SPECIFIC COURSE

```python
COURSES = ['Excel', 'Python']

X_LABEL = 'Nota'
Y_LABEL = 'Aluno'
COLOR_PALETTE = 'pastel'
HUE = 'Curso'
MARKER = 'o'
LEGEND = True

TITLE = f'Average grade of students per course - {' and '.join(COURSES)}'
```

```python
df_course = df[df['Curso'].isin(values=COURSES)]

sns.lineplot(data=df_course, x=X_LABEL, y=Y_LABEL, marker=MARKER, palette=COLOR_PALETTE, hue=HUE, legend=LEGEND)
plt.title(label=TITLE)
plt.grid(True)
plt.show()
```
