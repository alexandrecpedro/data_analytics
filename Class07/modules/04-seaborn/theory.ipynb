## 2. SIMPLE EXAMPLE

### 2.1. EXPLORE DATA ANALYSIS (EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt

DATASET = 'tips'
COLUMN_NAME = 'total_bill'
CHART_TITLE_LABEL = 'Distribution of the Total Bill Amount'
```

```python
tips = sns.load_dataset(name=DATASET)
```

```python
tips.info()
```

```python
tips.describe(include='all')
```

### 2.2. CHARTS

#### 2.2.1. TOTAL_BILL X TIP

Hypothesis: the higher the bill amount, the higher the tip amount

##### 2.2.1.1. SCATTERPLOT CHART

```python
X_FIELD = 'total_bill'
Y_FIELD = 'tip'
TITLE = 'Relation between Total Bill Amount and Tips'

sns.scatterplot(data=tips, x=X_FIELD, y=Y_FIELD)
plt.title(label=TITLE)
plt.show()
```

##### 2.2.1.2. LINEAR REGRESSION CHART

Display a trend line to represent the relation between the bill amount and the tip amount

```python
X_FIELD = 'total_bill'
Y_FIELD = 'tip'
ORDER = 1
TITLE = 'Relation between Total Bill Amount and Tips'

sns.regplot(data=tips, x=X_FIELD, y=Y_FIELD, order=ORDER)
plt.title(label=TITLE)
plt.show()
```

##### 2.2.1.3. LINEPLOT CHART

Display the relation between total bill amount and tip amount, by audience

```python
# SEX = 'Female'

X_FIELD = 'total_bill'
Y_FIELD = 'tip'
MARKER = 'o'
# COLOR_PALETTE = 'pastel'
HUE = 'sex'
LEGEND = True

TITLE = f'Relation between Total Bill Amount and Tips - {HUE} audience'
```

```python
# dataf = tips[tips.sex == SEX]

sns.lineplot(data=tips, x=X_FIELD, y=Y_FIELD, marker=MARKER, hue=HUE, legend=LEGEND)
plt.title(label=TITLE)
plt.show()
```

#### 2.2.2. TIP AMOUNT X DAYS

##### 2.2.2.1. BAR CHART

```python
X_FIELD = 'day'
Y_FIELD = 'tip'
ESTIMATOR = 'mean' ## Tip amount average
# ESTIMATOR = 'sum'
COLOR_PALETTE = 'pastel'
HUE = 'day'
LEGEND = False

TITLE = 'Relation between Tip Amount Average and Days'
```

```python
# ==============
# DEPRECATED VERSION
# ==============
# sns.barplot(data=tips, x=X_FIELD, y=Y_FIELD, estimator=ESTIMATOR, palette=COLOR_PALETTE)

sns.barplot(data=tips, x=X_FIELD, y=Y_FIELD, estimator=ESTIMATOR, palette=COLOR_PALETTE, hue=HUE, legend=LEGEND)

plt.title(label=TITLE)
plt.show()
```

##### 2.2.2.2. BOXPLOT CHART

```python
X_FIELD = 'day'
Y_FIELD = 'tip'
COLOR_PALETTE = 'pastel'
HUE = 'day'
LEGEND = False

TITLE = 'Distribution of Tip Amounts by Day'
```

```python
sns.boxplot(data=tips, x=X_FIELD, y=Y_FIELD, palette=COLOR_PALETTE, hue=HUE, legend=LEGEND)

plt.title(label=TITLE)
plt.show()
```

##### 2.2.2.3. HISTOGRAM CHART

```python
X_FIELD = 'tip'
BINS = 25
KIND = True
COLOR_PALETTE = 'orange'
LEGEND = False

TITLE = 'Histogram of Tip Amounts'
```

```python
sns.histplot(data=tips, x=X_FIELD, bins=BINS, kde=KIND, color=COLOR_PALETTE)
plt.title(label=TITLE)
plt.show()
```

##### 2.2.2.4. VIOLINPLOT CHART

```python
X_FIELD = 'day'
Y_FIELD = 'tip'
COLOR_PALETTE = 'pastel'
HUE = 'day'
LEGEND = False

TITLE = 'Distribution of Tip Amounts by Day'
```

```python
# ==============
# DEPRECATED VERSION
# ==============
# sns.violinplot(data=tips, x=X_FIELD, y=Y_FIELD, color=COLOR_PALETTE, hue=HUE, legend=LEGEND)

sns.violinplot(data=tips, x=X_FIELD, y=Y_FIELD, palette=COLOR_PALETTE, hue=HUE, legend=LEGEND)
plt.title(label=TITLE)
plt.show()
```
