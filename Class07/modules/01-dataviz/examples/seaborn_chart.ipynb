## 2. PLOT EXAMPLE

### 2.1. CONSTANTS

```python
DATASET = 'tips'
COLUMN_NAME = 'total_bill'
CHART_TITLE_LABEL = 'Distribution of the Total Bill Amount'
```

### 2.2. LOAD AN EXAMPLE DATASET

```python
import seaborn as sns

tips = sns.load_dataset(name=DATASET)
```

### 2.3. GENERATING AN HISTOGRAM OF A COLUMN

```python
import matplotlib.pyplot as plt

sns.histplot(data=tips[COLUMN_NAME], kde=True)
plt.title(CHART_TITLE_LABEL)
plt.show()
```
