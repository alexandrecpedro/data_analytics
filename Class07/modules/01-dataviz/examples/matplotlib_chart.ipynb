## 2. PLOT EXAMPLE

```python
import matplotlib.pyplot as plt

## LINE CHART
LINE_01 = [1, 2, 3, 4]
LINE_02 = [10, 20, 25, 30]
CHART_TITLE_LABEL = 'Example of chart with Matplotlib'
X_AXIS_LABEL = 'X Axis'
Y_AXIS_LABEL = 'Y Axis'

plt.plot(LINE_01, LINE_02)
plt.title(label=CHART_TITLE_LABEL)
plt.xlabel(xlabel=X_AXIS_LABEL)
plt.ylabel(ylabel=Y_AXIS_LABEL)
plt.show()
```
